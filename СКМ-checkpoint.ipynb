```python
import numpy as np 
from scipy import optimize 
import matplotlib.pyplot as plt 

def f(x): 
    return (x-1)**2 

x=[] 
iv=np.arange(2,300,0.25) 
for i in iv: 
    r=optimize.newton(f,i) 
    r=r-1 
    x.append(r) 

plt.plot(iv,x, '.b')

plt.xlabel('номер шага')
plt.ylabel('ошибка')
```

Теорема о сходимости метода Ньютона  
Пусть корень уравнения f(x) = 0 отделен на отрезке $[a;b]$ (функция f(x)
непрерывна на $[a;b]$ и на концах его принимает разные знаки), а производные
$f'(x)$ и $f''(x)$ отличны от нуля и сохраняют постоянные знаки на $[a;b]$. Тогда,
если выбрать начальное приближение х0 $[a;b]$ так, чтобы $f'(x0) ∙ f''(x0) > 0$, то
последовательность приближений, определяемая формулой
$xᵢ=xᵢ₋₁-f(xᵢ₋₁)/f'(xᵢ₋₁)$ сходится.
Вычислим скорость сходимости для каждого их уравнений $(x-1)²=0$ и $e²ˣ-2eˣ+1=0$
Скорость сходимости-степень приближения к точному решению на каждой итерации.  
Для первого уравнения:  
$f(x)=(x-1)²$  
$f'(x)=2x-2$  
$x-f(x)/f'(x)=x-(x-1)²/(2x-2)=(x-1)/2$  
По графику видно, что итерации сходятся линейно.

```python
import numpy as np 
import math
import matplotlib.pyplot as plt
%matplotlib inline

y=[]
x=np.arange(0,5,0.5)
def f(x): 
    return (x-1)/2 
for x1 in x:
    y.append(f(x1))

plt.plot(x,y,'.')    
```

Для второго уравненя:  
$f(x)=e²ˣ-2eˣ+1$  
$f(x)=2e²ˣ-2eˣ$  
$x-f(x)/f'(x)=x-(e²ˣ-2eˣ+1)/(2e²ˣ-2eˣ)=x-1/eˣ-1/2$
По графику видно, что итерации сходятся линейно.

```python
import numpy as np 
import math
import matplotlib.pyplot as plt
%matplotlib inline

y=[]
x=np.arange(0,5,0.5)
def f(x): 
    return x-(math.exp(x)-1)/(2*math.exp(x)) 
for x1 in x:
    y.append(f(x1))

plt.plot(x,y,'.')    
```
